```python
def predict_callback(patch: np.ndarray):
    """Receives a batch with size [tiles, window, window, channels]
    Returns a batch with size [tiles, window, window, classes]
    """
    res = []
    for i in range(patch.shape[0]):
        rgb = (np.random.rand(3) - 0.5)
        res.append(np.clip(patch[i] - rgb, 0, 1))
    return np.array(res)
```

```python
prediction = predict_smooth_windowing_v2(image=image, tile_size=128, subdivisions=4, num_classes=3, prediction_fn=predict_callback, batch_size=5, mirrored=True)
```
